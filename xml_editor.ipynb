```python
import cv2
import glob
import os
```

```python
path = "annotations"
base_img_path = os.path.join("..", "image")
list_file = glob.glob(path + "/*")
```

```python
import xml.etree.ElementTree as ET

# Get Each XML
for idx, xml_file in enumerate(list_file):
  mytree = ET.parse(xml_file)
  myroot = mytree.getroot()
  print(idx, xml_file)

  # Get Filenames
  for filename in myroot.iter('filename'):
    fname = filename.text
  
  for path in myroot.iter('path'):
    new_path = os.path.join(base_img_path, fname)
    path.text = new_path

  mytree.write(xml_file)
```
