```python
import pandas as pd

data = pd.read_csv("Food Ingredients and Recipe Dataset with Image Name Mapping.csv")
data = data[data['Ingredients'].str.len() > 2]

data
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from nltk.stem import PorterStemmer
#from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import re
import string

search = data["Ingredients"]
keywords = []

for txt in search :
    print("loading")
    #preprocessing
    #make all word to lowercase
    txt = txt.lower()
    #remove number
    rm_number = str.maketrans('', '', string.digits)
    txt = txt.translate(rm_number)
    #remove punctuations
    rm_punctuations = str.maketrans('', '', string.punctuation)
    txt = txt.translate(rm_punctuations)
    #remove UNICODE characters
    txt = txt.encode("ascii", "ignore")
    txt = txt.decode()
    #remove whitespaces
    txt = txt.strip()
            
    #factory = StopWordRemoverFactory()
    #stopword = factory.create_stop_word_remover()
    #txt = stopword.remove(txt)
    #print(text)
    
    #Tokenizing
    tokenize = word_tokenize(txt)
    
    #Filtering
    filtered = []
    stopword = stopwords.words('english')
    new_stopword = ["tbsp", "lb", "tsp", "oz", "g", "inch", "inches", "tablespoons", "cup", "teaspoon"]
    stopword.extend(new_stopword)
    
    for t in tokenize :
        if t not in stopword :
            filtered.append(t)
    tokenize = filtered
    
    #Stemming
    stemmer = PorterStemmer()
    new_keyword = []
    
    for word in tokenize:
        new_keyword.append(stemmer.stem(word))
    keywords.append(new_keyword)
    #keywords.append(tokenize)
```

```python
keywords
```

```python
#analyzing
from nltk.probability import FreqDist

scores = []
for keyword in keywords:
    tf = FreqDist(keyword)
    most_common = tf.most_common()[0]
    filtered_freq = []
    
    for word in tf.most_common() :
        #if(word[1] >= most_common[1]/2) :
        filtered_freq.append(word)
    
    scores.append(filtered_freq)
    
scores
```

```python
#open file
file = open("text_result.py", "w")

#convert variable to string
str = repr(scores)
file.write("scores = " + str + "\n")

#close file
file.close()
```
